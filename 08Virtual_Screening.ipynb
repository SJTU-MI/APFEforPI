```python
import pandas as pd
import joblib
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
import warnings
warnings.filterwarnings("ignore")
```

```python
model_path="./pretrained_model/"
RF_name = 'RF_Opt.model'
KRR_name = 'KRR_Opt.model'
MLP_name = 'MLP_Opt.model'
RF_model = joblib.load(model_path+RF_name)
KRR_model = joblib.load(model_path+KRR_name)
MLP_model = joblib.load(model_path+MLP_name)
```

```python
path="./dataset/"
file='Cor_train.pkl'
train_df= pd.read_pickle(path+file)
select_feature=[
'Monomer_length',
 'MW_ratio',
 'Mass_max',
 'Mass_ave',
 'K_bond_ave',
 'K_ang_ave',
 'nH',
 'AATS0d',
 'AATS0Z',
 'ATSC2Z',
 'ATSC5Z',
 'AATSC0Z',
 'AATSC2Z',
 'BCUTd-1h',
 'BCUTZ-1h',
 'nHBDon',
 'MIC0',
 'MIC1',
 'Kier3',
 'Mor02',
 'Mor14',
 'Mor19',
 'SMR_VSA1',
 'SMR_VSA3',
 'VSA_EState4']
train_select=train_df[select_feature]
train_select=train_df[select_feature]
```

```python
PolyInfo_file='DatasetB.pkl'
PolyInfo_df= pd.read_pickle(path+PolyInfo_file)
polyinfo_select=PolyInfo_df[select_feature]
polyinfo_select
```

```python
PI_file='DatasetC.pkl'
PI_df= pd.read_pickle(path+PI_file)
PI_select=PI_df[select_feature]
PI_select
```

```python
for i in [train_select,polyinfo_select,PI_select]:
    i.index = range(i.shape[0])
scaler.fit(train_select)
polyinfo_data=scaler.transform(polyinfo_select)
PI_data=scaler.transform(PI_select)
```

```python
#PolyInfo database
polyinfo_RF = RF_model.predict(polyinfo_data)
polyinfo_KRR = KRR_model.predict(polyinfo_data)
polyinfo_MLP = MLP_model.predict(polyinfo_data)

#PI database
PI_RF = RF_model.predict(PI_data)
PI_KRR = KRR_model.predict(PI_data)
PI_MLP = MLP_model.predict(PI_data)
```

```python
polyinfo_TC = pd.DataFrame({'RF':polyinfo_RF,'KRR':polyinfo_KRR,'MLP':polyinfo_MLP})
polyinfo_TC
```

```python
PI_TC = pd.DataFrame({'RF':PI_RF,'KRR':PI_KRR,'MLP':PI_MLP})
PI_TC
```
